```python
# -----------------------------
# My Python Pet Game üê∂üê±
# -----------------------------

# The pet has "stats" (like happiness, energy, hunger)
# We store them inside a dictionary so we can easily look them up by name
pet_stats = {
    "happiness": 5,  # how happy the pet feels (0 = sad, 10 = super happy)
    "energy": 5,     # how much energy the pet has (0 = very tired, 10 = full of energy)
    "hunger": 5      # how hungry the pet is (0 = full, 10 = starving)
}

# The player‚Äôs "inventory" is a list of items they can use
inventory = ["apple", "ball"]


# -----------------------------
# FUNCTIONS (things our game can do)
# -----------------------------

def show_stats():
    """Show the pet's current stats on the screen"""
    print("\nPet Stats:")
    print("Happiness:", pet_stats["happiness"])
    print("Energy:", pet_stats["energy"])
    print("Hunger:", pet_stats["hunger"])


def feed_pet():
    """Feed the pet if you have food in your inventory"""
    if "apple" in inventory:  # check if we have an apple
        print("You feed your pet an apple. Yum! üçé")
        pet_stats["hunger"] -= 2     # decrease hunger (pet is less hungry)
        inventory.remove("apple")    # remove the apple from the bag (used up)
    else:
        print("You have no food! Find some!")
    check_mood()  # always check the pet's mood after doing something


def play_with_pet():
    """Play fetch with the pet if you have a ball"""
    if "ball" in inventory:
        print("You play fetch with your pet. Fun! üéæ")
        pet_stats["happiness"] += 2   # pet gets happier
        pet_stats["energy"] -= 1      # playing uses energy
    else:
        print("You have no toy to play with!")
    check_mood()


def put_pet_to_sleep():
    """Let the pet take a nap"""
    print("Your pet takes a nap... üò¥")
    pet_stats["energy"] += 3   # pet gains energy
    pet_stats["hunger"] -= 1   # sleeping makes it a bit hungrier
    check_mood()


def check_mood():
    """Check the pet's mood and make sure stats stay between 0 and 10"""

    # Make sure no stat goes below 0 or above 10
    for key in pet_stats:
        if pet_stats[key] < 0:
            pet_stats[key] = 0
        elif pet_stats[key] > 10:
            pet_stats[key] = 10

    # Show mood based on happiness stat
    if pet_stats["happiness"] >= 7:
        print("Your pet is very happy! üòä")
    elif pet_stats["happiness"] >= 4:
        print("Your pet is okay. üòê")
    else:
        print("Your pet is sad. üò¢")


# -----------------------------
# MAIN GAME LOOP
# -----------------------------

def start_game():
    """This is the main part of the game that keeps running until you quit"""
    print("Welcome to My Python Pet! üêæ")

    # Keep looping forever until the player chooses to quit
    while True:
        show_stats()  # show how the pet is doing

        # Show menu options
        print("\nWhat would you like to do?")
        print("1. Feed Pet")
        print("2. Play with Pet")
        print("3. Put Pet to Sleep")
        print("4. Quit Game")

        # Ask the player to type a choice
        choice = input("Choose 1-4: ")

        # Call the right function based on what they typed
        if choice == "1":
            feed_pet()
        elif choice == "2":
            play_with_pet()
        elif choice == "3":
            put_pet_to_sleep()
        elif choice == "4":
            print("Goodbye! Thanks for taking care of your pet! üëã")
            break   # "break" ends the while loop ‚Üí game over
        else:
            print("Please choose a number from 1-4.")


# -----------------------------
# START THE GAME
# -----------------------------
start_game()
```
